***E1***. Grafique en 3D la superficie de la función
$$
    f(x,y) = \dfrac{\sin\left(\sqrt{x^{2} + y^{2}}\right)}{\sqrt{x^{2} + y^{2}}}
$$
en el dominio $-10\leq x,y\leq 10$.

**Solución**: Establecemos la ecuación de la gráfica, juntos con sus dominios y el número de puntos, asi como sus etiquetas para una mayor comprensión

````python
# Script
fig = plt.figure()
ax = Axes3D(fig)
x = np.linspace(-10,10,50)
y = np.linspace(-10,10,80)
X,Y = np.meshgrid(x,y)
Z = (np.sin((x**2+Y**2)**(1/2)))/((x**2+Y**2)**(1/2))
ax.plot_surface(X, Y, Z,cmap='jet')
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.set_zlabel("$f(x,y)$")
ax.set_title("$f(x,y) = \dfrac{\sin(\sqrt{x^2+y^2})}{\sqrt{x^2+y^2}}$")
plt.show()
````

***E2***. Haga un *script* que pida al usuario la expresión de una función $f(x)$. Usando la expresión para $f(x)$ calcule su derivada y haga la gráfica de $f(x)$ y $f'(x)$ en los mismos ejes. Para hacer el cálculo de la derivada utilice cálculo simbólico. Para la gráfica es necesario que utilice diferentes estilos para las curvas de la función y su derivada, además de poner etiquetas a los ejes y una leyenda. **Solución**: Solicitamos la funcion y calculamos la derivada con el modulo SymPy para después graficar ambar ecuaciones

````python
# Script
x = sympy.symbols("x")
expr = sympy.sympify(input('Dame una expresión: '))
pt = int(input("Dame el numero de puntos a graficar: "))
i1 = int(input("Dame el valor del limite izquierdo: "))
i2 = int(input("Dame el valor del limite derecho: "))
der = sympy.diff(expr)
import numpy as np
import matplotlib.pyplot as plt
f = sympy.lambdify(x,expr)
g = sympy.lambdify(x,der)
e = np.linspace(i1,i2,pt)
plt.plot(e,f(e),"-r",label= expr)
plt.plot(e,g(e),"ob",label= der)
ax = plt.gca() 
ax.set_xlabel("$x$")
ax.set_ylabel("$f(x)$") 
ax.set_title("Función dada y su derivada")
plt.show()
````

***E3***. La manera de hacer graficas en coordenadas polares con ```matplolib``` es usar la función ```polar``` de ```pyplot```. La función ```polar``` recibe como argumentos de entrada el ángulo y el radio. Por ejemplo, para graficar la curva *cardioide* cuya ecuación en coordenadas polares está dada por

$$
    r(\theta) = a(1 + \cos\theta)
$$

donde $\theta$ es el ángulo y $a>0$, se utiliza el siguiente código (de forma análoga a usar ```plot```)

````python
# Ejemplo de la grafica de la cardioide usando coordenadas polares
a = 4 
t = np.linspace(0,2*np.pi) # angulo
r = a*(1 + np.cos(t)) # radio
plt.polar(t,r,'--',color = 'firebrick',lw=2)
plt.show()
````

Considere la ecuación para la *rosa polar* en coordenadas polares

$$
    r(\theta) = \left|a\sin\left(\dfrac{k}{2}\theta + \phi\right)\right|.
$$

Utilizando esta ecuación, cree una función que haga la gráfica de esta curva. Como nombre de la función use ```rosaPolar``` y como argumentos de entrada:
* ```a``` (```float```), define la longitud de los pétalos.
* ```k``` (```float```), define el número de pétalos.
* ```phi``` (```float```), define un ángulo de rotación $\phi$.
* ```lim``` (```float```), define el intervalo para el ángulo $\theta$, esto es, $0\leq\theta\leq$```lim```.
* ```p``` (```int```), define el número de puntos a utilizar para dibujar la gráfica.

La función sólo hace la gráfica en coordenadas polares, no regresa ningún valor u objeto.

Después haga un *script* para pedir al usuario los valores de los argumentos de entrada de la función ```rosaPolar```. Pida el ángulo de rotación ```phi``` en grados en lugar de radianes. Use el manejo de excepciones para mandar un mensaje al usuario cuando haya cometido un error al definir los valores, esto para evitar la interrupción del *script*.

**Solución**: Definimos una función donde trabajaremos toda la ecuacion de la rosa polar para despues solicitar los valores de las variables y mostrar la grafica polar en funcion a los valores dados

````python
# Funciones
def rosaPolar(a,k,phi,lim,p):
    t = np.linspace(0,lim,p)
    r = a*np.sin((k/2)*t+phi) 
    plt.polar(t,r,'--',color = 'firebrick',lw=1)
    plt.show()
````

````python
# Script
try:
    a = float(input("Define la longitud de los petalos: ")) 
    k = float(input("Define la el número de petalos: ")) 
    phi = float(input("Define el ángulo de rotación en grados: ")) 
    lim = float(input("Define el intervalo del ángulo theta: ")) 
    p = int(input("Define el número de puntos para dibujar la gráfica: "))
except:
    print("Error en los argumentos de entrada")
rosaPolar(a,k,phi,lim,p)
````
