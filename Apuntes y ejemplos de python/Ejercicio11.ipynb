***E1***. Cree una función que determine si un numero entero positivo es primo o no. Para el nombre de la función use ```esprimo``` y está debe recibir como argumento de entrada un número entero (```int```) y como argumento de salida debe regresar un booleano (```bool```).

Posteriormente, en un *script* cree la lista ```L``` con los números del 2 al 1000 y utilice la función ```filter``` junto con su función para crear una segunda lista ```P``` con los primos contenidos en ```L```. Al final del *script*, use un ciclo para mostrar los elementos en ```P``` de la siguiente manera:
* ```P(1) = 2```
* ```P(2) = 3```
* ```P(3) = 5```
* ...

**Solución**: Definimos un rango con el cual el programa revisara si un numero eso no divisble entre uno número distinto de cero y de si mismo, para descartar cualquier numero no primo, posteriormente lo filtramos para obtener la lista deseada

````python
# Funciones
def esprimo(num):
    for n in range(2, floor(sqrt(num))+1):
        if num % n == 0:
            return False
    return True
````

````python
# Script
L = np.arange(2,1000)
P = filter(lambda L: esprimo(L) == True,L)
for z,k in enumerate(list(P),1):
    print(f"P({z}) = {k}")
````

**Solución**: Para la primer parte, escribimos una funcion que nos permita conocer si un año es bisiesto o no, posteriormente descartamos los casos falsos para agilizar el programa. Para la segunda parte definimos la funcion con los nombres de cada mes para despues evaluarlos con .split en nuestra función.

````python
# Funciones
#Funcion para validar si un año es bisiesto
def bisiesto(año):
    if año < -400:
        if año%400 == 0:
            return True
        elif año%100 == 0:
            return False
        elif año%4 == 0:
            return True
        elif año%4 > 0:
            return False

    elif año <= 400:
        if año%400 == 0:
            return False 
        elif año%100 == 0:
            return False
        elif año%4 == 0:
            return True
        elif año%4 > 0:
            return False
            
    elif año > 400:
        if año%400 == 0:
            return True
        elif año%100 == 0:
            return False
        elif año%4 == 0:
            return True
        elif año%4 > 0:
            return False
#Funcion para validar una fecha
def validaFecha(dia,mes,año):
    if dia == 31 and (mes == 4 or mes == 6 or mes == 9 or mes ==11):
        return False
    elif bisiesto(año) == True and (dia >= 30 and mes == 2):
        return False
    elif bisiesto(año) == False and (dia>=29 and mes == 2):
        return False
    return True
#Funcion para el cambio de nombre
def Cambiaformato(mes):
    if mes == 1:
        return "Enero"
    elif mes == 2:
        return "Febrero"
    elif mes == 3:
        return "Marzo"
    elif mes == 4:
        return "Abril"
    elif mes == 5:
        return "Mayo"
    elif mes == 6:
        return "Junio"
    elif mes == 7:
        return "Julio"
    elif mes == 8:
        return "Agosto"
    elif mes == 9:
        return "Septiembre"
    elif mes == 10:
        return "Octubre"
    elif mes == 11:
        return "Noviembre"
    elif mes == 12:
        return "Diciembre"
````

````python
# Script
d = eval(input("Dame un dia: "))
m = eval(input("Dame un mes: "))
a = eval(input("Dame un año: "))
if validaFecha(d,m,a) == True:
    print("La fecha ingresada si existe")

else:
    print("La fecha ingresada no existe") 
````

````python
#Script cambio de formato
fecha = input("Ingrese una fecha en formato dd-mm-aaaa: ")
f = fecha.split("-")
L = list(map(int, f))
mm = Cambiaformato(L[1])
if validaFecha(L[0],L[1],L[2]) == True:
    print(f"La fecha ingresada es: {L[0]} de {mm} de {L[2]}")
else:
    print("La fecha ingresada no existe")
````
